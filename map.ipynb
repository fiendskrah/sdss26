```python
#| label: fig:event-map
import folium

lat, lon = 37.872014, -122.267505

m = folium.Map(
    location=[lat, lon],
    zoom_start=17,
    tiles="OpenStreetMap",
)

folium.Marker(
    [lat, lon],
    popup="Spatial Data Summit",
    tooltip="AI Futures Lab @ WeWork 2120 University Ave, 7th Floor, Berkeley, CA 94704",
    icon=folium.Icon(icon="info-sign"),
).add_to(m)

m
```
